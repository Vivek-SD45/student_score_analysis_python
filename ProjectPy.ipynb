```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# drop unnamed Coulumn


```python
df= df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender") #plot the chart for gender
ax.bar_label(ax.containers[0])  #use for labelling the chart
plt.title("Gender Distribution")
plt.show()
```

### from the above chart we have analysis that:
### the number of female in data is more then the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore": "mean","WritingScore":"mean"})
print(gb)
```

```python

plt.figure(figsize=(8,6))
sns.heatmap(gb , annot = True)
plt.title("Relationship between parent's education and student's score")

plt.show()
```

#### from the above chart we conclude that the education of parents have good impact on there score

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore": "mean","WritingScore":"mean"})
print(gb1)
```

```python

plt.figure(figsize=(8,8))
sns.heatmap(gb1 , annot = True)
plt.title("Relationship between parent's merital status and student's score")

plt.show()
```

### from the above chart we conclude that the ParentMaritalStatus  of parents have negligible or no impact on student score

# Distribution of  Ethnic Groups

```python
groupA=df.loc[(df['EthnicGroup']== "group A")].count()
groupB=df.loc[(df['EthnicGroup']== "group B")].count()
groupC=df.loc[(df['EthnicGroup']== "group C")].count()
groupD=df.loc[(df['EthnicGroup']== "group D")].count()
groupE=df.loc[(df['EthnicGroup']== "group E")].count()
print(groupA["EthnicGroup"])
# by this we got count of for ethinic grp in group A
l=["group A","group B","group C","group D","group E"] #to make label
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnics  Groups")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='WklyStudyHours',y='MathScore')
plt.title('Box Plot Outlier for Study Hours')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='MathScore', y='ReadingScore', data=df, hue='Gender', palette='coolwarm')
plt.title('Math Score vs Reading Score')
plt.show()
```

```python
plt.figure(figsize=(8,6))
correlation_matrix = df[['MathScore', 'ReadingScore', 'WritingScore']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Correlation between Scores')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='EthnicGroup', data=df, palette='viridis')
plt.title('Count of Students by Ethnic Group')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='ParentEduc', y='MathScore', data=df, palette='Paired')
plt.title('Average Math Score by Parent Education Level')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='TestPrep', y='WritingScore', data=df, palette='muted')
plt.title('Writing Scores by Test Preparation Status')
plt.show()
```

```python
sns.pairplot(df[['MathScore', 'ReadingScore', 'WritingScore', 'Gender']], hue='Gender', palette='husl')
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.boxplot(x='LunchType', y='ReadingScore', hue='TestPrep', data=df, palette='Set3')
plt.title('Reading Scores by Lunch Type and Test Preparation Status')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='TransportMeans', hue='Gender', data=df, palette='Set1')
plt.title('Transport Means by Gender')
plt.show()
```

```python
df['NrSiblings'] = df['NrSiblings'].fillna(0)  # Handle missing values in 'NrSiblings'
plt.figure(figsize=(8,6))
sns.barplot(x='NrSiblings', y='MathScore', data=df, palette='cubehelix')
plt.title('Average Math Score by Number of Siblings')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
sns.jointplot(x='MathScore', y='WritingScore', data=df, kind='reg', color='g')
plt.title('Math vs Writing Scores with Regression Line')
plt.show()
```
